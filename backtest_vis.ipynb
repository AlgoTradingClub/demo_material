```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bars = pd.read_pickle('bars.pkl')
portfolio = pd.read_pickle('portfolio.pkl')
ratings = pd.read_pickle('ratings.pkl')
```

```python
def vis(symbol, all_bars, all_trades):
    bars = all_bars.loc[symbol]
    
    trades = all_trades.loc[all_trades['symbol'] == symbol]
    buys = trades.loc[trades['qty'] > 0]
    sells = trades.loc[trades['qty'] < 0]
    
    sns.lineplot(x=bars.index, y=bars['close']).set_title('Trades for ' + symbol)
    sns.scatterplot(x='time', y='price', data=buys, color='green', s=100, label='BOUGHT')
    sns.scatterplot(x='time', y='price', data=sells, color='red', s=100, label='SOLD')
```

```python
symbol = ratings.sort_values('rating', ascending=False).iloc[14]['symbol']
```

```python
plt.figure(figsize=(16, 10))
vis(symbol, bars, portfolio)
plt.savefig('crvs.png')
```
